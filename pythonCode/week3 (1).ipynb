```python
# Data for X and Y
X = [6.3, 2.5, 8.9, 3.5, 9.5, 2.3, 2.1, 3.4, 5.1]
Y = [50, 71, 25, 92, 82, 15, 2, 60, 31]

# Ranging Y values with respect to X by dividing Y by X
Y_ranged = [y / x for x, y in zip(X, Y)]

# Bar plot
plt.bar(X, Y_ranged, color='b',width=.5)
plt.title('Bar Graph')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()

# Scatter plot
plt.scatter(X, Y_ranged, color='r')
plt.title('Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

```python
# Select two features for scatter plotting (sepal_length and sepal_width)
X = file_path['sepal_length']
Y = file_path['sepal_width']

# Compute the IQR for Sepal Length and Sepal Width
Q1_X = X.quantile(0.25)
Q3_X = X.quantile(0.75)
IQR_X = Q3_X - Q1_X

Q1_Y = Y.quantile(0.25)
Q3_Y = Y.quantile(0.75)
IQR_Y = Q3_Y - Q1_Y

# Define outliers as values that fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers_X = (X < (Q1_X - 1.5 * IQR_X)) | (X > (Q3_X + 1.5 * IQR_X))
outliers_Y = (Y < (Q1_Y - 1.5 * IQR_Y)) | (Y > (Q3_Y + 1.5 * IQR_Y))

# Combine outliers for either X or Y
outliers = outliers_X | outliers_Y

# Scatter plot
plt.scatter(X, Y, label='Data Points', color='green')

# Mark outliers in red
plt.scatter(X[outliers], Y[outliers], label='Outliers', color='red', marker='o')

plt.title('Scatter Plot of Sepal Length vs Sepal Width with Outliers Marked (IQR Method)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
