```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/ml dataset/iris.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
x=df.drop('species',axis=1)
x
```

```python
l=LabelEncoder()
y=l.fit_transform(df['species'])
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lgrg= LogisticRegression()
lgrg.fit(x_train,y_train)
y_train_prediction=lgrg.predict(x_train)
y_train_prediction
```

```python
lgrg.fit(x_train,y_train)
lgrg.score(x_train,y_train) *100
```

```python
lgrg.score(x_test,y_test) *100
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_accuracy = accuracy_score(y_train_prediction, y_train)
print('Accuracy on Training data : ', training_data_accuracy*100)
```

```python
y_test_prediction=lgrg.predict(x_test)
testing_data_accuracy = accuracy_score(y_test_prediction, y_test)
print('Accuracy on Testing data : ', testing_data_accuracy*100)
```

```python
!pip install pandas matplotlib scikit-learn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Group data by species and calculate the mean for each feature
species_means = df.groupby(iris.target).mean()

# Create the bar chart
species_means.plot(kind='bar', figsize=(10, 6))
plt.title('Average Measurements of Iris Species')
plt.xlabel('Species (0: Setosa, 1: Versicolor, 2: Virginica)')
plt.ylabel('Measurement (cm)')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(iris.feature_names)
plt.show
```
