```python
dff = pd.read_csv("/content/drive/MyDrive/ml dataset/heart.csv")
dff
```

```python
x = dff.drop("target", axis=1)
x
```

```python
y = dff["target"]
y
```

```python
dff.corr()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_fitures = SelectKBest(score_func=f_classif)
fit_fitures.fit(x, y)
```

```python
fs =pd.DataFrame(fit_fitures.scores_, index=x.columns,columns=['score values'])
fs
```

```python
top_9 = fs.nlargest(9, 'score values')
```

```python
print(top_9['score values'])
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
model.fit(x, y)
```

```python
fs = pd.DataFrame(model.feature_importances_, index=x.columns, columns=['score values'])
fs
```
