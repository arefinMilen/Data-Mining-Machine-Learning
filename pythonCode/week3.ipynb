```python

import matplotlib.pyplot as plt
import numpy as np


x = [6.3, 2.5, 8.9, 3.5, 9.5, 2.3, 2.1, 3.4, 5.1]
y = [50, 71, 25, 92, 82, 15, 2, 60, 31]


sorted_indices = np.argsort(x)
x_sorted = np.array(x)[sorted_indices]
y_sorted = np.array(y)[sorted_indices]


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(x_sorted, y_sorted, color='blue', width=0.3)
plt.title('Bar Graph of y vs. x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.xticks(x_sorted)


plt.subplot(1, 2, 2)
plt.scatter(x_sorted, y_sorted, color='red')
plt.title('Scatter Plot of y vs. x')
plt.xlabel('x values')
plt.ylabel('y values')
plt.xticks(x_sorted)


plt.tight_layout()
plt.show()
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/ml dataset/iris.csv")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(0)
data_size = 100
x = np.random.normal(loc=5, scale=2, size=data_size)
y = np.random.normal(loc=50, scale=10, size=data_size)


df = pd.DataFrame({'x': x, 'y': y})


Q1 = df['y'].quantile(0.25)
Q3 = df['y'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_filtered = df[(df['y'] >= lower_bound) & (df['y'] <= upper_bound)]


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='x', y='y', color='red')
plt.title('Scatter Plot of Irish Data (Outliers Masked)')
plt.xlabel('X-values')
plt.ylabel('Y-values')
plt.grid(True)


plt.show()

```
